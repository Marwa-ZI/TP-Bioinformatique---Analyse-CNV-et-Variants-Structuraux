### Contexte clinique
Le gene **FLT3** (FMS-Like Tyrosine kinase 3) est frequemment mute dans les leucemies aigues myeloblastiques (LAM).
Les duplications internes en tandem (**FLT3-ITD**) touchent les exons 14-15 et constituent un marqueur pronostique defavorable important.

FiLT3r detecte ces variants par une approche basee sur les **k-mers** : il compare les reads au genome de reference
et identifie ceux portant des sequences aberrantes caracteristiques d'une duplication.

**Arguments a identifier :**
- `-k` : taille des k-mers (obligatoire)
- `--sequences` : fichiers FASTQ d'entree (separes par une virgule, sans espace)
- `--ref` : fichier de reference FASTA
- `--out` : prefixe de sortie
- `--vcf` : generer un fichier VCF en plus du JSON
