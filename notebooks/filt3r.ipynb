## 3. Lancement de l'analyse

Parametres utilises :
- `-k 12` : k-mers de taille 12 (bon compromis sensibilite/specificite)
- `--ref` : sequence des exons 14-15 de FLT3 (hg38)
- `--sequences` : reads R1 et R2 en paired-end
- `--out results` : prefixe de sortie
- `--vcf` : export en format VCF

```python
! /usr/local/bin/filt3r \
    --ref flt3_exon14_15_hg38.fa \
    -k 12 \
    --sequences flt3_dna_R1.fastq.gz,flt3_dna_R2.fastq.gz \
    --out results \
    --vcf
```
